 **SOFIFA WEBSITE SCRAPING**

```python
#Importing libraries
import pandas as pd
import re
import requests
import numpy as np
from bs4 import BeautifulSoup
a = [0, 60, 120, 180, 240, 300, 360, 420, 480, 540]
NAME = []
AGE = []
OVA = []
POT = []
TEAM_NAME = []
CONTRACT = []
VALUE = []
WAGE = []
TOTAL_STATS = []
   
for i in a:
    
        url = f'https://sofifa.com/players?offset={i}'
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content,'html')
        all_name=soup.find(("table",{"class":"table table-hover persist-area"})) 
        
        
        for body in all_name.findAll(("tbody",{"class":"list"})) : #to print name
            rows=body.findAll('tr')
            for row in rows :
              player_name=row.find("a",{"class":"tooltip"}).div.text
              NAME.append(re.sub(r'\xa0|20','',player_name))   #removing unwanted part

              player_age=row.find("td",{"class":"col col-ae"}).text
              AGE.append(player_age) 

              player_ova=row.find("td",{"class":"col col-oa"}).span.text
              OVA.append(player_ova)   

              player_pot=row.find("td",{"class":"col col-pt"}).span.text
              POT.append(player_pot) 
                   
              player_team=row.findAll("td",{"class":"col-name"})[1].a.text
              TEAM_NAME.append(player_team)

              player_contract=row.find("div",{"class":"sub"}).text
              CONTRACT.append(re.sub(r'\n','',player_contract)) 

              player_wage=row.find("td",{"class":"col col-wg"}).text
              WAGE.append(re.sub(r'€|K','',player_wage))
            
              player_value=row.find("td",{"class":"col col-vl"}).text
              VALUE.append(re.sub(r'€|M','',player_value)) 

              player_stats=row.find("td",{"class":"col col-tt"}).span.text
              TOTAL_STATS.append(player_stats) 

  

df = pd.DataFrame({"Name": NAME,
                   "  Age  ": AGE,
                   "Overall_Rating ": OVA,
                   " Potential": POT,
                   "Team_Name": TEAM_NAME,
                   "Contract_Year": CONTRACT,        #Creating dataFrame
                   "Value(M) ": VALUE,
                   "  Wage(K)": WAGE,
                   "Total_Stats": TOTAL_STATS})



df.index = np.arange(1,len(df)+1)
 # Check
df.to_csv('fifa.csv') # Export the data  
df.head(150).style.set_table_styles([{'selector': 'th',
  'props': [('background', '#204A74  '), 
            ('color', '#E1F5F9'), ('font-wight', 'bold ') ,('padding' ,'10px' ),('text-align' ,'center' )]},
 
 {'selector': 'td',
  'props': [('background', '#F4F6FC '), 
            ('color', '#204A74  '),('padding' ,'10px' ),('text-align' ,'center' )]},
]) 
```
